# Pandas Exercise: Time Series (Apple Stock)

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/Time_Series/Apple_Stock/appl_1980_2014.csv)

### Step 3. Assign it to a variable apple

```python
apple=pd.read_csv("appl_1980_2014.csv")
apple
```

### Step 4.  Check out the type of the columns

```python
apple.dtypes
```

### Step 5. Transform the Date column as a datetime type

```python
apple['Date']=pd.to_datetime(apple['Date'])
apple.dtypes
```

### Step 6.  Set the date as the index

```python
apple.set_index('Date',inplace=True)
apple
```

### Step 7.  Is there any duplicate dates?

```python
#apple.index[apple.index.duplicated()].unique()
apple.index.duplicated(keep=True)
```

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
apple1=apple.sort_index()
```

```python
apple1
```

### Step 9. Get the last business day of each month

```python
apple_monthly=apple.resample('M').sum()
apple_monthly.index
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
apple1.index[-1]-apple1.index[0]
```

### Step 11.  How many months in the data we have?

```python
apple_monthly=apple.resample('M').sum()
len(apple_monthly)
```

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [13.5,9]
plt.plot(apple1.index,apple1['Adj Close'])
plt.show()
```
