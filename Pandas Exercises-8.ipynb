### Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK

```python
df2=pd.DataFrame()
df=retail_data.groupby('Country').sum()['Quantity'].sort_values(ascending=False).head(11)
df.drop(df.index[0],inplace=True)
df2['countries']=df.index
Clist=df2['countries'].tolist()
Clist
retail_hist=retail_data[retail_data['Country'].isin(Clist)]
de1=pd.DataFrame()
de1=(retail_hist.groupby(['Quantity','Country'])).count()
de1['Count']=de1['InvoiceNo']
de1.drop(['InvoiceNo','StockCode','Description','InvoiceDate','UnitPrice','CustomerID'],axis=1,inplace=True)
de1
de1[['Count']].plot(kind='hist',bins=40,rwidth=1.2,figsize=(18,6))
plt.show()
```

### Step 6. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries

```python
retail_data.columns
```

```python
retail_data['Quantity_per_unit_price']=retail_data['Quantity']/retail_data['UnitPrice']
retail_data
```

```python
retail = retail_data[retail_data['UnitPrice']!=0]
groups=retail.groupby(['CustomerID','Country']).sum()['Quantity_per_unit_price']
groups=groups.round(decimals=0)
groups=groups.sort_values(ascending=False)
```

```python
group=pd.DataFrame(groups)
group=group.reset_index()
group1=group[group['Country'].isin(['Netherlands','United Kingdom','EIRE'])]
group1
```

```python
colors={'Netherlands':'red','United Kingdom':'blue','EIRE':'yellow'}
fig,ax=plt.subplots()
ax.scatter(group1['CustomerID'],group1['Quantity_per_unit_price'],c=group1['Country'].apply(lambda x: colors[x]))
plt.show()
```

```python
import seaborn as sns
retail=group1
sns.lmplot(x='CustomerID',y='Quantity_per_unit_price',data=retail,hue='Country',fit_reg=False)
```
